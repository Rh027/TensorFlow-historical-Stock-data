```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.metrics import mean_squared_error
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split

```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/!ripple-dat - 0.csv')
df.head()
```

```python
X = df.drop(['timeOpen','timeLow', 'timeClose','timeHigh','timeHigh','priceHigh','volume'],axis=1)
y=df['priceHigh']
```

```python
X.head()
```

```python
y.head()
```

```python
X['MEAN'] = X.mean(axis=1)
X_t = X.drop(['priceOpen','priceLow','priceClose'],axis=1)
X_t.head()
```

```python
Xmean_train, Xmean_test, y_train, y_test = train_test_split(X_t, y, test_size=0.2, random_state=42)
```

```python
model = Sequential()
model.add(Dense(units=32, activation='relu', input_dim=len(Xmean_train.columns)))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=1))
```

```python
model.compile(loss='mse', optimizer=Adam(learning_rate=0.01), metrics=['accuracy'])
```

```python
model.fit(Xmean_train, y_train, epochs=200, batch_size=32, validation_data=(Xmean_test, y_test))
```

```python
y_hat = model.predict(Xmean_test)
```

```python
y_hat
```

```python
mean_squared_error(y_test, y_hat)
```
